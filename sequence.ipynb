```python
import pandas as pd

# Sample DataFrame
data = {
    'CODES': ['A01'],
    'PRODUCTS': ['Pen : 2015 - 2020'],
    'UNITS': ['Box: 14, 19'],
    'PRICE': [15],
    'REMARKS': ['']
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract the range of years from the PRODUCTS column
product = df['PRODUCTS'].iloc[0]

# Split the string using multiple separators (':', '-')
parts = product.replace(':', '-').split('-')

if len(parts) >= 3:
    start_year = int(parts[1].strip())
    end_year = int(parts[2].strip())

    # Create sequences for PRODUCTS and UNITS columns
    sequence = [f'Pen {year}' for year in range(start_year, end_year + 1)] * 2
    units_sequence = [f'Box: {", ".join(parts[-1].split(","))}' for year in range(start_year, end_year + 1)] * 2

    # Update the original DataFrame by repeating the rows and updating PRODUCTS and UNITS columns
    new_length = len(sequence)
    df = pd.concat([df] * (new_length // len(df)), ignore_index=True)
    df['PRODUCTS'] = sequence[:len(df)]
    df['UNITS'] = units_sequence[:len(df)]

# Display the updated DataFrame vertically
df
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'CODES': ['A01'],
    'PRODUCTS': ['Pen : 2015 - 2020'],
    'UNITS': ['Box: 14, 19'],
    'PRICE': [15],
    'REMARKS': ['']
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract the range of years from the PRODUCTS column
product = df['PRODUCTS'].iloc[0]

# Split the string using multiple separators (':', '-')
parts = product.replace(':', '-').split('-')

if len(parts) >= 3:
    start_year = int(parts[1].strip())
    end_year = int(parts[2].strip())

    # Create sequences for PRODUCTS and UNITS columns
    sequence = [f'Pen {year}' for year in range(start_year, end_year + 1)] * 2
    unit_sequence = ['Box 14' if idx < len(sequence) / 2 else 'Box 19' for idx in range(len(sequence))]

    # Update the original DataFrame by repeating the rows and updating PRODUCTS and UNITS columns
    df = pd.concat([df] * (len(sequence) // len(df)), ignore_index=True)
    df['PRODUCTS'] = sequence
    df['UNITS'] = unit_sequence

# Display the updated DataFrame vertically
df
```
